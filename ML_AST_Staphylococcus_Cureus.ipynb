```python
import pandas as pd
import numpy as np
import umap
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Ast_phd.csv', index_col=0)

# Extract drug sensitivity values (excluding drug names)
data_values = df.values  # Convert dataframe to a NumPy array

# Impute NaN values with the mean of each column
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean') # Replace NaN with the mean of the column
data_values = imputer.fit_transform(data_values)

# Apply UMAP for dimensionality reduction
reducer = umap.UMAP(n_components=2, random_state=42)
embedding = reducer.fit_transform(data_values)

# Convert embedding to a DataFrame
umap_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
umap_df['Drug'] = df.index  # Assign drug names as labels

# Plot UMAP clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=umap_df, x='UMAP1', y='UMAP2', hue='Drug', palette='tab10', legend=False)

# Add labels for clarity
for i, drug in enumerate(umap_df['Drug']):
    plt.text(umap_df['UMAP1'][i], umap_df['UMAP2'][i], drug, fontsize=9, alpha=0.7)

plt.title('UMAP Clustering of Drug Sensitivity')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.show()
```

```python
import pandas as pd
import numpy as np
import umap
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/Ast_phd.csv', index_col=0)  # First column is drug names

# Transpose so that drugs are columns (features), and samples are rows
df_transposed = df.T

# Impute NaN values with the mean of each column (before applying UMAP)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Replace NaN with the mean of the column
df_transposed_imputed = imputer.fit_transform(df_transposed) # Apply imputation to df_transposed

# Apply UMAP for dimensionality reduction (keeping all drugs)
reducer = umap.UMAP(n_components=2, random_state=42)
# Use the imputed data for UMAP
embedding = reducer.fit_transform(df_transposed_imputed)

# Convert embedding to DataFrame
umap_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
umap_df['Drug'] = df_transposed.index  # Assign drug names

# Unique colors for each drug
plt.figure(figsize=(10, 7))
palette = sns.color_palette('tab10', n_colors=len(umap_df['Drug'].unique()))

# Scatter plot with different colors for each drug
sns.scatterplot(data=umap_df, x='UMAP1', y='UMAP2', hue='Drug', palette=palette)

# Add title and labels
plt.title('UMAP Clustering of Drugs')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.legend(title='Drugs', bbox_to_anchor=(1, 1))  # Move legend outside

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file
df = pd.read_csv('/content/Ast_phd.csv')

# Remove NaN values
df = df.dropna()

# Extract drug names (1st row) and their values
drugs = df.columns.tolist()

# Create a new DataFrame in long format
data_list = []
for drug in drugs:
    values = df[drug].values  # Get 150 values per drug
    for val in values:
        # Add small random jitter for UMAP-like spread effect
        x_jitter = np.random.normal(0, 0.3)
        y_jitter = np.random.normal(0, 0.3)
        data_list.append([drug, val, x_jitter, y_jitter])

# Convert to DataFrame
data_df = pd.DataFrame(data_list, columns=['Drug', 'Cluster', 'X_Jitter', 'Y_Jitter'])

# Map cluster values to numeric positions
cluster_mapping = {1: 'Sensitive', 2: 'Intermediate', 3: 'Resistant'}
data_df['Cluster_Label'] = data_df['Cluster'].map(cluster_mapping)

# Convert Drug names into numeric values
drug_labels = {drug: i for i, drug in enumerate(drugs)}
data_df['Drug_X'] = data_df['Drug'].map(drug_labels) + data_df['X_Jitter']  # Apply jitter for better spread
data_df['Cluster_Y'] = data_df['Cluster'] + data_df['Y_Jitter']  # Apply jitter to Y for better separation

# Define color palette for drugs
palette = sns.color_palette('husl', len(drugs))

# Plot UMAP-like scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data_df, x='Drug_X', y='Cluster_Y', hue='Drug', palette=palette, alpha=0.8, edgecolor=None, s=50)

# Adjust labels
plt.xticks(ticks=list(drug_labels.values()), labels=drugs, rotation=90)  # X-axis with drug names
plt.yticks([1, 2, 3], ['Sensitive', 'Intermediate', 'Resistant'])  # Cluster categories
plt.xlabel('Drugs')
plt.ylabel('Clusters')
plt.title('UMAP-Like Clustering of Drug Sensitivity')

# Show legend outside
plt.legend(title='Drugs', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/content/Ast_phd.csv', header=0)

# Convert data: Mark resistance (3) as 1, others as 0
binary_df = df.replace({1: 0, 2: 0, 3: 1})  # Only resistance counts

# Create co-occurrence matrix (drug-drug resistance pattern)
co_occurrence = np.dot(binary_df.T, binary_df)  # Dot product for co-resistance

# Convert to DataFrame
co_matrix = pd.DataFrame(co_occurrence, index=df.columns, columns=df.columns)
np.fill_diagonal(co_matrix.values, 0)  # Remove self-loops

# Create Network Graph
G = nx.Graph()

# Add nodes (drugs)
for drug in co_matrix.columns:
    G.add_node(drug)

# Add edges based on co-resistance strength
for i in range(len(co_matrix.columns)):
    for j in range(i + 1, len(co_matrix.columns)):
        if co_matrix.iloc[i, j] > 0:  # Add edge if co-resistance exists
            G.add_edge(co_matrix.index[i], co_matrix.columns[j], weight=co_matrix.iloc[i, j])

# Plot the Network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Set layout

# Draw nodes & edges
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', edge_color='gray', font_size=10, font_weight='bold')
edges = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edges, font_size=8)

plt.title("Co-Resistive Pattern Network")
plt.show()
```

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/content/Ast_phd.csv', header=0)

# Convert data: Mark resistance (3) as 1, others as 0
binary_df = df.replace({1: 0, 2: 0, 3: 1})  # Only resistance counts

# Create co-occurrence matrix
co_occurrence = np.dot(binary_df.T, binary_df)
co_matrix = pd.DataFrame(co_occurrence, index=df.columns, columns=df.columns)
np.fill_diagonal(co_matrix.values, 0)

# Create Network Graph
G = nx.Graph()

# Add nodes with colors
colors = plt.cm.rainbow(np.linspace(0, 1, len(co_matrix.columns)))
node_color_map = dict(zip(co_matrix.columns, colors))

for drug in co_matrix.columns:
    G.add_node(drug, color=node_color_map[drug])

# Add edges based on co-resistance strength
for i in range(len(co_matrix.columns)):
    for j in range(i + 1, len(co_matrix.columns)):
        weight = co_matrix.iloc[i, j]
        if weight > 0:
            G.add_edge(co_matrix.index[i], co_matrix.columns[j], weight=weight)

# Improve visualization
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, k=0.3, seed=42)  # Better node spacing

# Draw nodes with colors
node_colors = [G.nodes[n]['color'] for n in G.nodes]
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=800, alpha=0.8)

# Draw edges with varying thickness
edges = nx.get_edge_attributes(G, 'weight')
edge_widths = [edges[e] / 10 for e in edges]  # Scale edge thickness
nx.draw_networkx_edges(G, pos, edgelist=edges.keys(), width=edge_widths, edge_color='gray', alpha=0.6)

# Draw labels with offset
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold', verticalalignment='bottom')

plt.title("Improved Co-Resistive Network", fontsize=14)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/Ast_phd.csv', header=0)

# Convert resistance states (Ensure values: 1=Sensitive, 2=Intermediate, 3=Resistant)
df.replace({1: "Sensitive", 2: "Intermediate", 3: "Resistant"}, inplace=True)

# Create an empty transition matrix
states = ["Sensitive", "Intermediate", "Resistant"]
transition_matrix = pd.DataFrame(0, index=states, columns=states)

# Calculate transitions for each drug (column-wise)
for col in df.columns:
    previous_state = df[col].shift(1)  # Shift data to compare consecutive values
    transitions = list(zip(previous_state, df[col]))  # Pair previous & current state

    # Count occurrences of each transition
    for prev, curr in transitions:
        if prev in states and curr in states:
            transition_matrix.loc[prev, curr] += 1

# Normalize the transition matrix to probabilities
transition_matrix = transition_matrix.div(transition_matrix.sum(axis=1), axis=0).fillna(0)

# Heatmap visualization
plt.figure(figsize=(8, 6))
sns.heatmap(transition_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5,
            xticklabels=states, yticklabels=states)

plt.xlabel("Next Resistance State")
plt.ylabel("Previous Resistance State")
plt.title("Drug Resistance Transition Matrix Heatmap")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx

# Load Data
df = pd.read_csv("Ast_phd.csv", header=0)

# Convert Resistance Values: 1 = Sensitive, 2 = Intermediate, 3 = Resistant
df_binary = df.replace({1: 0, 2: 0, 3: 1})  # 1 means resistant, 0 otherwise

# Compute Co-Occurrence Matrix (Transition Counts)
transition_matrix = np.dot(df_binary.T, df_binary)
transition_df = pd.DataFrame(transition_matrix, index=df.columns, columns=df.columns)

# Normalize Each Row (Convert Counts to Probabilities)
transition_df = transition_df.div(transition_df.sum(axis=1), axis=0).fillna(0)

# Plot Heatmap of Transition Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(transition_df, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Markov Chain Transition Matrix (Resistance Spread)")
plt.xlabel("Next Resistance")
plt.ylabel("Current Resistance")
plt.show()

# Create Markov Chain Graph
G = nx.DiGraph()
for i in transition_df.index:
    for j in transition_df.columns:
        if transition_df.loc[i, j] > 0:  # Only add transitions with probability > 0
            G.add_edge(i, j, weight=transition_df.loc[i, j])

# Draw Graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
edges = G.edges(data=True)
edge_weights = [d['weight'] * 5 for (u, v, d) in edges]  # Scale edge width

nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", width=edge_weights, node_size=2000)
plt.title("Markov Chain Model: Drug Resistance Transition")
plt.show()

# Predict Next-Step Resistance Distribution (Matrix Squaring)
next_step_resistance = np.linalg.matrix_power(transition_df.values, 2)
next_step_df = pd.DataFrame(next_step_resistance, index=df.columns, columns=df.columns)

# Plot Future Predictions
plt.figure(figsize=(12, 8))
sns.heatmap(next_step_df, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Predicted Resistance Transition (Next Step)")
plt.xlabel("Next Resistance")
plt.ylabel("Current Resistance")
plt.show()

# Compute Steady-State Distribution (Long-Term Prediction)
_, eigenvectors = np.linalg.eig(transition_df.T)
steady_state = eigenvectors[:, 0].real
steady_state /= steady_state.sum()
steady_df = pd.DataFrame(steady_state, index=df.columns, columns=["Steady-State Probability"])

# Plot Steady-State Distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=steady_df.index, y=steady_df["Steady-State Probability"], palette="viridis")
plt.title("Steady-State Drug Resistance Probability")
plt.xlabel("Drug")
plt.ylabel("Probability of Resistance in Long Run")
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib.cm as cm

# Load Data
df = pd.read_csv("Ast_phd.csv", header=0)

# Convert Resistance Values: 1 = Sensitive, 2 = Intermediate, 3 = Resistant
df_binary = df.replace({1: 0, 2: 0, 3: 1})  # 1 means resistant, 0 otherwise

# Compute Co-Occurrence Matrix (Transition Counts)
transition_matrix = np.dot(df_binary.T, df_binary)
transition_df = pd.DataFrame(transition_matrix, index=df.columns, columns=df.columns)

# Normalize Each Row (Convert Counts to Probabilities)
transition_df = transition_df.div(transition_df.sum(axis=1), axis=0).fillna(0)

# Create Markov Chain Graph
G = nx.DiGraph()
for i in transition_df.index:
    for j in transition_df.columns:
        if transition_df.loc[i, j] > 0:  # Only add transitions with probability > 0
            G.add_edge(i, j, weight=transition_df.loc[i, j])

# Generate Pastel Colors for Nodes (Drugs)
colors = cm.Pastel1(np.linspace(0, 1, len(df.columns)))

# Assign colors to nodes in the graph
node_color_map = {drug: color for drug, color in zip(df.columns, colors)}

# Draw Graph with Pastel Colors
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
edges = G.edges(data=True)
edge_weights = [d['weight'] * 5 for (u, v, d) in edges]  # Scale edge width

# Draw nodes with pastel colors
node_colors = [node_color_map[node] for node in G.nodes]
nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color="gray", width=edge_weights, node_size=2000)

plt.title("Markov Chain Model: Drug Resistance Transition (Pastel Colors)")
plt.show()
```
